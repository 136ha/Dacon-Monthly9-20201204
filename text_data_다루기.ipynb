```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv("train.csv", encoding = 'utf-8')
print("Number of train texts: ", train.shape[0])
```

```python
# 클렌징 - 앞뒤 공백 제거할때 쓰임
train['text_strip']  = train['text'].str.strip()  # 앞 뒤 공백을 제거
train['text_lstrip'] = train['text'].str.lstrip() # 앞 공백을 제거
train['text_rstrip'] = train['text'].str.rstrip() # 뒤 공백을 제거
```

```python
# 패딩, 고정길이 데이터를 이용해야 하는 경우 왼쪽 오른쪽 중앙에 놓고 패딩
train['text_pad']    = train['text'].str.pad(width=20, side='left', fillchar='_') # 지정길이 패딩
train['text_center'] = train['text'].str.center(width=20, fillchar='_') # 중앙놓고 지정길이 패딩
train['text_ljust']  = train['text'].str.ljust(width=20, fillchar='_')  # 왼쪽놓고 지정길이 패딩
train['text_rjust']  = train['text'].str.rjust(width=20, fillchar='_')  # 오른쪽놓고 지정길이 패딩
train['text_zfill']  = train['text'].str.zfill(width=20)                # 지정길이 0패딩
```

```python
# 소문자, 대문자, 앞자리만 대문자, 소문자 대문자 뒤집기
train['text_lower']      = train['text'].str.lower()      # 모두 소문자로 변경
train['text_upper']      = train['text'].str.upper()      # 모두 대문자로 변경
train['text_capitalize'] = train['text'].str.capitalize() # 앞문자 대문자로 변경
train['text_title']      = train['text'].str.title()      # 단위별 앞문자 대문자로 변경
train['text_swapcase']   = train['text'].str.swapcase()   # 소문자는 대문자, 대문자는 소문자로 변경 

# 입력 패턴 또는 글자를 대체, 예제에서는 .을 _로 변경
train['text_replace']    = train['text'].str.replace(pat='.', repl='_', regex=False)
```

```python
# 문자열의 위치 찾기
train['text_find']    = train['text'].str.find(sub='.')           # 왼쪽부터 sub값 검색후 위치반환
train['text_findall'] = train['text'].str.findall(pat='[a-zA-Z]') # 찾은 모든 값 반환
train['text_rfind']   = train['text'].str.rfind(sub='.')          # 오른쪽부터 sub값 검색후 위치반환
# train['text_index']   = train['text'].str.index(sub='.')          # 왼쪽부터 sub값 검색후 위치반환
# train['text_rindex']  = train['text'].str.rindex(sub='.')         # 오른쪽부터 sub값 검색후 위치반환
```

```python
# 문자열 길이 측정
train['text_len']   = train['text'].str.len()              # 길이 반환
train['text_count'] = train['text'].str.count(pat='[0-9]') # 문자열 중 패턴에 일치한 수 반환
```

```python
# 숫자, 알파벳, 소문자, 대문자 등을 확인
train['text_isalnum']   = train['text'].str.isalnum()   # 알파벳 또는 숫자로만 구성 여부
train['text_isalpha']   = train['text'].str.isalpha()   # 알파벳으로만 구성 여부
train['text_isdecimal'] = train['text'].str.isdecimal() # 숫자문자로만 구성 여부
train['text_isdigit']   = train['text'].str.isdigit()   # 숫자문자로만 구성 여부
train['text_islower']   = train['text'].str.islower()   # 소문자로만 구성 여부
train['text_isnumeric'] = train['text'].str.isnumeric() # 숫자문자로만 구성 여부
train['text_isspace']   = train['text'].str.isspace()   # 공백(Whitespace)으로만 구성 여부
train['text_istitle']   = train['text'].str.istitle()   # TitleCase형태로 구성 여부
train['text_isupper']   = train['text'].str.isupper()   # 대문자로만 구성 여부
```

```python
# 문자열의 시작값 종료값 또는 특정 패턴 찾기
train['text_startswith'] = train['text'].str.startswith(pat='h')     # 좌측값이 입력패턴과 일치 여부
train['text_endswith']   = train['text'].str.endswith(pat='com')     # 우측값이 입력패턴과 일치 여부
train['text_contains']   = train['text'].str.contains(pat='kr', regex=False) # 값 중 패턴포함 여부
train['text_match']      = train['text'].str.match(pat='[a-zA-Z@.]') # 입력패턴과 일치 여부
```
